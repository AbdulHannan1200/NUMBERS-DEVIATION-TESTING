```python
# generate random integer values
from random import seed
from random import randint
# seed random number generator
seed(1)
temp_lst = list()
# generate some integers
for i in range(15000):
    value = randint(0, 36)
    temp_lst.append(value)
    print(i,value)
    
new_lst = list()
for index in range(len(temp_lst)):
    if index == len(temp_lst)-1:
        break
    else:
        temp_var_01 = temp_lst[index]
        temp_var_02= temp_lst[index+1]
        tup = list([temp_var_01,temp_var_02])
        new_lst.append(tup)
    
print(new_lst)
```

```python
import pandas as pd
# =     
import numpy as np
```

```python
df = pd.DataFrame()
for ind in range(len(new_lst)):
    df.loc[ind,'X'] = new_lst[ind][0]
    df.loc[ind,'y'] = new_lst[ind][1]
```

```python
cols = ['X', 'y']
df[cols] = df[cols].applymap(np.int64)
```

```python
Y = pd.get_dummies(df['y']).values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,16))
df.plot(x="X", y="y", kind="bar")
```

```python
from sklearn.model_selection import train_test_split
X=df[['X']].values
y= df['y'].values
#scaled_features
#target
#X_train, X_test, y_train, y_test = train_test_split(scaled_features,target, test_size=0.25, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=42)
```

```python
print(X_test[0])
print(y_test[0])
```

```python
import tensorflow as tf
model=tf.keras.models.Sequential()
from tensorflow.keras.utils import to_categorical
```

```python
model.add(tf.keras.layers.Dense(6,input_dim=1,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(16,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(56,activation=tf.nn.relu))
model.add(tf.keras.layers.Dense(37,activation=tf.nn.softmax))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
val_loss,val_accuracy=model.evaluate(X_test,y_test)
print('Validation Loss:',val_loss)
print('Validation Accuracy:',val_accuracy)
```

```python
embed_dim = 128
lstm_out = 196
max_fatures=37
model_lstm = Sequential()
model_lstm.add(Embedding(max_fatures, embed_dim,input_length = X.shape[1]))
model_lstm.add(LSTM(lstm_out, dropout=0.2, recurrent_dropout=0.2))
model_lstm.add(Dense(37,activation='softmax'))
model_lstm.compile(loss = 'categorical_crossentropy', optimizer='adam',metrics = ['accuracy'])
print(model_lstm.summary())
```

```python
history = model_lstm.fit(X_train,  y_train, 
                    batch_size=20, epochs=150)
```

```python
print("The original list is : " + str(output_lst)) 
  
# initialize N  
N = 6
  
# Indices of N largest elements in list 
# using sorted() + lambda + list slicing 
res = sorted(range(len(output_lst)), key = lambda sub: output_lst[sub])[-N:] 
  
# printing result 
print("Indices list of max N elements is : " + str(res))
```

```python
def Make_Tuple(res,output_lst):
    tuple_lst = list()
    for tup in range(len(res)):
        inner = res[tup]
        outer = output_lst[inner]
        tuple_lst.append([inner,outer])
        
    return tuple_lst
        
```

```python
def Predict_Classes(number):
    output_lst = model_lstm.predict([number])
    output_lst = list(output_lst[0])
    
    print("The original list is : " + str(output_lst)) 
  
    N = 6
    
    res = sorted(range(len(output_lst)), key = lambda sub: output_lst[sub])[-N:] 

    print("Indices list of max N elements is : " + str(res))
    
    result_tuples = Make_Tuple(res,output_lst)
    
    return result_tuples
    
```

```python
def Plot_Graph(output):
    X_graph = list()
    y_graph = list()
    for i in range(len(output)):
        X_graph.append(output[i][0])
        y_graph.append(output[i][1])
        
    plt.scatter(X_graph,y_graph)
    plt.show()
        
    
```
